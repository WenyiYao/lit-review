```python
import pdfplumber

def extract_text_from_pdf(pdf_path):
    with pdfplumber.open(pdf_path) as pdf:
        text = ''
        for page in pdf.pages:
            text += page.extract_text()
    return text

# Extract text from a PDF file
pdf_text = extract_text_from_pdf('Walker_2022_34656053.pdf')
```

```python
pdf_text
```

```python
import spacy

# Load a spaCy model with biomedical NER capabilities
nlp = spacy.load("en_core_sci_sm")

# Process the extracted text
doc = nlp(pdf_text)

# Define the target variant
target_variant = "c.489_490delTG"

# Extract sentences containing the target variant
sentences_with_variant = [sent.text for sent in doc.sents if target_variant in sent.text]

# Print the extracted sentences
for sentence in sentences_with_variant:
    print(sentence)
```
